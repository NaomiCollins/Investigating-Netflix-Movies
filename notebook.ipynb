Performed exploratory data analysis on the netflix_data.csv dataset to explore movies from the 1990s decade.

Filtered the data for movies released in the 1990s and identified the most frequent movie duration, saving an approximate value as an integer called duration.

Defined short movies as those under 90 minutes and counted the number of short action movies released in the 1990s, saving this count as short_movie_count.

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Keep only movies
movies_df = netflix_df[netflix_df['type'] == 'Movie']

# Filter for release years between 1990 and 1999
movies_1990s = movies_df[(movies_df['release_year'] >= 1990) &
                          (movies_df['release_year'] <= 1999)]

# Find the most frequent duration
duration = int(movies_1990s['duration'].mode()[0])

# Count short action movies (< 90 minutes)
short_action_movies = movies_1990s[
    (movies_1990s['genre'] == 'Action') &
    (movies_1990s['duration'] < 90)
]

short_movie_count = short_action_movies.shape[0]

print(short_movie_count)
```
